    #Data Frame

```python
data1 = pd.Series({'courses': 'mathematics', 'Number_of_required' : 12, 'Dept_prof_number': 15})
data2 = pd.Series({'courses': 'statistics', 'Number_of_required' : 11, 'Dept_prof_number': 9})
data3 = pd.Series({'courses': 'data_analytics', 'Number_of_required' : 15, 'Dept_prof_number': 12})

data = pd.DataFrame([data1,data2,data3], index = ['un1','un2','un3'])
data
```

```python
#Loc and iloc in dataframe 
#we can use loc only if we have assigned index in dataframe 
data.loc['un1']
```

```python
data.iloc[0]
```

```python
data.iloc[0]['courses']
```

```python
#Transpose of the data
data.T
```

```python
#Drop data
data.drop('un1')
```
