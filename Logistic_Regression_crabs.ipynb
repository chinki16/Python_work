```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
data.groupby('GoodSpine').mean()
```

```python
#First converting GoodSpine and Dark to dummy variables 
data['Dark'] = data['Dark'].map({'yes': 1, 'no': 0})
data['GoodSpine'] = data['GoodSpine'].map({'yes':1 , 'no' : 0})
data
```

```python
#Running Logistic Regression
import statsmodels.api as sm
data_y = data['GoodSpine']
data_x = data.drop(['GoodSpine'],axis = 1)
logit = LogisticRegression()
model = sm.Logit( data_y, data_x)
result = model.fit()
result.summary()
```

```python
#Model fitting
logit = LogisticRegression()
logit.fit(data_x, data_y)
pred_y= logit.predict(data_x)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(data_y, pred_y)
print(confusion_matrix)
```
